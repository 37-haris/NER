# 1. Test the small pipeline

```python
nlp = spacy.load("en_core_web_sm")
nlp
```

```python
# test the pipeline
doc = nlp("Apple is looking at buying U.K. startup for $1 billion")

for token in doc:
    print("Token:",token.text, "Lemma:", token.lemma_, "POS:", token.pos_, token.tag_, token.dep_,
            token.shape_, token.is_alpha, token.is_stop)
```

Using trained pipeline from spaCy, you can extract linguistic features like token, lemma, Part of Speech (POS), POS tag, dependency, etc.

Now, let's use this spaCy pipeline for NER in the following texts:

```python
# Text 1
doc1 = nlp("Victor Marie Hugo was a French poet, playwright, novelist, statesman and human rights activist.")
doc1
```

```python
from spacy import displacy # displacy helps us visualizing the entities

# Text 1
displacy.render(doc1, style="ent", jupyter=True)
```

```python
# Text 2
doc2 = nlp("Pablo Ruiz Picasso was a Spanish painter, sculptor, printmaker, ceramicist, and theatre designer who spent most of his adult life in France.")
doc2
```

```python
# Text 2
displacy.render(doc2, style="ent", jupyter=True)
```

# 2. Text the large pipeline

```python
nlp = spacy.load("en_core_web_lg")
nlp
```

```python
doc = nlp("Pablo Ruiz Picasso was a Spanish painter, sculptor, printmaker, ceramicist, and theatre designer who spent most of his adult life in France.")
doc
```

```python
type(doc)
```

```python
# Check the recognized entities from the text
doc.ents
```

```python
doc.ents[0], type(doc.ents[0])
```

```python
from spacy import displacy
displacy.render(doc, style="ent", jupyter=True)
```

```python
for ent in doc.ents:
    print(ent.text, ent.start_char, ent.end_char, ent.label_)
```

Entity description:

Value: 'Pablo Ruiz Picasso' --> Label: 'Person' --> Value position in text: (start:0, end:18)
Value: 'Spanish' --> Label: 'NORP' --> Value position in text: (start:25, end:32)
Value: 'France' --> Label: 'GPE' --> Value position in text: (start:133, end:139)
